```python
import pandas as pd
from sklearn.cluster import KMeans
import math

# === 1. Load data ===
df = pd.read_excel("Employee Location Details Data Only Location.xlsx")
df.columns = df.columns.str.strip()  # clean column names

# === 2. Drop rows with missing GPS data ===
df.dropna(subset=['latitude', 'longitude'], inplace=True)

# === 3. Loop through each shift (A, B, C, D) ===
shifts = ['A', 'B', 'C', 'D']

for active_shift in shifts:
    print(f"\n========== Processing Shift {active_shift} ==========")

    # Merge G and H into current shift
    df['Effective Group'] = df['Shift'].apply(
        lambda x: active_shift if x in ['G', 'H'] else x
    )

    # Filter relevant employees
    shift_df = df[df['Effective Group'] == active_shift].copy()

    # Check how many clusters (max 10 per cluster)
    total_employees = len(shift_df)
    if total_employees == 0:
        print(f"No employees for shift {active_shift}")
        continue

    n_clusters = math.ceil(total_employees / 10)
    print(f"Total Employees: {total_employees}, Required Clusters: {n_clusters}")

    if total_employees < n_clusters:
        print(f"Skipping shift {active_shift} (too few employees)")
        continue

    # === 4. Apply KMeans clustering ===
    coords = shift_df[['latitude', 'longitude']]
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    shift_df['cluster'] = kmeans.fit_predict(coords)

    # === 5. Print clusters ===
    for cluster_num in sorted(shift_df['cluster'].unique()):
        cluster_data = shift_df[shift_df['cluster'] == cluster_num]
        print(f"\nShift {active_shift} - Cluster {cluster_num + 1}:")
        print(f"Total Points: {len(cluster_data)}")
        for _, row in cluster_data.iterrows():
            print(f"   {row['Emp Name']:<25} {row['latitude']:.6f}   {row['longitude']:.6f}")

        # Check if cluster exceeds 10
        if len(cluster_data) > 10:
            print(" WARNING: Cluster exceeds 10 employees!")

    # === 6. Save to Excel ===
    shift_df.to_excel(f"Clustered_Shift_{active_shift}.xlsx", index=False)
```

```python
import pandas as pd
from sklearn.cluster import KMeans
import math

def cluster_with_capacity(df_points, capacity=10, cluster_label_offset=0):
    """
    Recursively cluster df_points so that each cluster has <= capacity employees.
    cluster_label_offset used to number clusters uniquely across recursive calls.
    Returns a DataFrame with a 'cluster' column with cluster numbers.
    """
    total_points = len(df_points)
    if total_points <= capacity:
        df_points['cluster'] = cluster_label_offset
        return df_points

    n_clusters = math.ceil(total_points / capacity)
    coords = df_points[['latitude', 'longitude']]
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    df_points['cluster_temp'] = kmeans.fit_predict(coords)

    result_df = pd.DataFrame()
    current_label = cluster_label_offset

    for cluster_id in range(n_clusters):
        subcluster = df_points[df_points['cluster_temp'] == cluster_id].copy()
        if len(subcluster) <= capacity:
            subcluster['cluster'] = current_label
            current_label += 1
            result_df = pd.concat([result_df, subcluster])
        else:
            # Recursive split
            split_df = cluster_with_capacity(subcluster.drop(columns=['cluster_temp']), capacity, current_label)
            current_label = split_df['cluster'].max() + 1
            result_df = pd.concat([result_df, split_df])

    return result_df.drop(columns=['cluster_temp'])

# Load your data
df = pd.read_excel("Employee Location Details Data Only Location.xlsx")
df.columns = df.columns.str.strip()
df.dropna(subset=['latitude', 'longitude'], inplace=True)

shifts = ['A', 'B', 'C', 'D']

for active_shift in shifts:
    print(f"\n========== Processing Shift {active_shift} ==========")

    df['Effective Group'] = df['Shift'].apply(lambda x: active_shift if x in ['G', 'H'] else x)
    shift_df = df[df['Effective Group'] == active_shift].copy()

    total_employees = len(shift_df)
    if total_employees == 0:
        print(f"No employees for shift {active_shift}")
        continue

    print(f"Total Employees: {total_employees}")

    clustered_df = cluster_with_capacity(shift_df, capacity=10, cluster_label_offset=0)

    # Print clusters
    for cluster_num in sorted(clustered_df['cluster'].unique()):
        cluster_data = clustered_df[clustered_df['cluster'] == cluster_num]
        print(f"\nShift {active_shift} - Cluster {cluster_num + 1}:")
        print(f"Total Points: {len(cluster_data)}")
        for _, row in cluster_data.iterrows():
            print(f"   {row['Emp Name']:<25} {row['latitude']:.6f}   {row['longitude']:.6f}")

    # Save to Excel
    clustered_df.to_excel(f"Clustered_StrictCapacity_Shift_{active_shift}.xlsx", index=False)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns  # optional, for better colors
from sklearn.cluster import KMeans
import math

def cluster_with_capacity(df_points, capacity=10, cluster_label_offset=0):
    total_points = len(df_points)
    if total_points <= capacity:
        df_points['cluster'] = cluster_label_offset
        return df_points

    n_clusters = math.ceil(total_points / capacity)
    coords = df_points[['latitude', 'longitude']]
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    df_points['cluster_temp'] = kmeans.fit_predict(coords)

    result_df = pd.DataFrame()
    current_label = cluster_label_offset

    for cluster_id in range(n_clusters):
        subcluster = df_points[df_points['cluster_temp'] == cluster_id].copy()
        if len(subcluster) <= capacity:
            subcluster['cluster'] = current_label
            current_label += 1
            result_df = pd.concat([result_df, subcluster])
        else:
            split_df = cluster_with_capacity(subcluster.drop(columns=['cluster_temp']), capacity, current_label)
            current_label = split_df['cluster'].max() + 1
            result_df = pd.concat([result_df, split_df])

    return result_df.drop(columns=['cluster_temp'])

# Load data and preprocess
df = pd.read_excel("Employee Location Details Data Only Location.xlsx")
df.columns = df.columns.str.strip()
df.dropna(subset=['latitude', 'longitude'], inplace=True)
df = df.drop_duplicates(subset=['Emp Name'])

shifts = ['A', 'B', 'C', 'D']

for active_shift in shifts:
    print(f"\nProcessing Shift {active_shift}")

    df['Effective Group'] = df['Shift'].apply(lambda x: active_shift if x in ['G', 'H'] else x)
    shift_df = df[df['Effective Group'] == active_shift].copy()

    if len(shift_df) == 0:
        print(f"No data for Shift {active_shift}, skipping plot.")
        continue

    clustered_df = cluster_with_capacity(shift_df, capacity=10)

    # Plot clusters
    plt.figure(figsize=(10, 8))
    sns.scatterplot(
        data=clustered_df,
        x='longitude', y='latitude',
        hue='cluster',
        palette='tab20',
        legend='full',
        s=50,
        alpha=0.7
    )
    plt.title(f"Shift {active_shift} - Employee Clusters")
    plt.xlabel("Longitude")
    plt.ylabel("Latitude")
    plt.legend(title='Cluster', bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.tight_layout()
    plt.savefig(f"Shift_{active_shift}_clusters.png")
    plt.close()
    print(f"Saved Shift_{active_shift}_clusters.png")
```
